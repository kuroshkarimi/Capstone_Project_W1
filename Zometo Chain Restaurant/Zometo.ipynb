**Analyzing Zomato data**

Zometo is a chain restaurant, The data regarding which is attached as "Zometo.csv" to this directory. 

Task1: Compare the online and offline delivery of the restaurants- Which one is higher?
Task2: What types of restaurants are most popular with the general public?
Task3: What price range do couples prefer when dining out and in, separately?

Tthe required libraries are imported.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Zomato data .csv')
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize = (3,2))
sns.countplot(x = df['online_order'])
plt.title("Number of on-line and offline orders")
plt.grid()
```

```python
plt.figure(figsize = (4, 2))
sns.countplot(x = df['listed_in(type)'])
plt.title("Number of preferred restaurants with general public")
plt.xlabel('different type of restaurants')
plt.grid('True')
```

```python
plt.figure(figsize = (3, 3))
sns.boxplot(x = 'online_order', y = 'approx_cost(for two people)', data = df)
plt.title('Approximate cost range for online and offline orders')
plt.grid()
```
